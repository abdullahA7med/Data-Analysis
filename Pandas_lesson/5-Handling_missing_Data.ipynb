1) ### dropna()

- **jop of it -> drop Row/column which have null(NaN) or any missing data**

Drop the Rows that have NaN.

```python
df.dropna() # here droped all rows, which have NaN.
```

**dropna, have some parameters**

- **Like**
    - axis -> remove that column which have NaN => we use axis, like in numpy
        - axis=0 (defoult) => delete rows.
        - axis=1           => delete column.

    - subset()

```python
df.dropna(axis=1) # delet column

df.dropna(axis='columns') # delet column

# delete any column have NaN.
```

2) ### fillna()


**Jop of it**

- **when you need to fill NaN by specific value.**

```python

# will fill any NaN in dataFrame by entry value. 

df1 = df.fillna(0)
```

```python
df1.loc[df1['Style'] == 0]
```

```python
df1.loc[df1['Style'] == 0]

# now all rows in style column (which have null) will equal 0.
```

**What if i need**

- **in Top Ten column -> i need make Null values equla specific value (ex => 1, -1, 0, 'Not Available')**
- **and in column of Style**

```python
df
```

```python

# first call fillna, inner it creat python dictionary.
# The key of dic will be    -> The Name of column.
# The value of dic will be  -> the val you want.

# reassign in new var. to store changes.
df2 = df.fillna({"Style": 'Not available',
          "Top Ten" : -1})


# now the Null value in style column will be -> Not available
# and the Null value in Top Ten column will be -> -1
df2
```

```python
# To see Not available row, which we change it


df2.loc[df2['Style'] == 'Not available']
```

### if you looked at result üëÜ mybe noticed, 'Unrated', (we know how remove it) | lets look the sec way of to day lesson.

1) **make that value -> Unrated <- Null (NaN) | then remove it by dropna().**

- **To make it Null, we use replace function**

data[].replace('old_val','New_val')


```python
df['Stars'] = df['Stars'].replace('Unrated',np.nan)

df['Stars'].value_counts() # to check -> unrated <- was removed by Null
```

```python
# to see Null values
df[df['Stars'].isnull()]
```
