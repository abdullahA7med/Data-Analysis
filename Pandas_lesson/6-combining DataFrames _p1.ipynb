#### Note:
**there are 4 types to combining the data**

1) inner joining
2) outer joining
3) left joining
4) right joining

### 1. inner joining :

**To use it -> must found 1 column shered bettwen them. and sorted by same order in two data frame**

the elemint, which not be same order -> will delet it.

- in our case -> item_category_id, which shered.



```python
# 
# the function which do that is -> merge(df1, df2 , how=joining_type, on=column_join_by-it (shered column))
```

```python
a = pd.DataFrame({'cat_name':['Elec','Games','Kitchen'],'cat_id':[0,1,2]})

a
```

```python
b=pd.DataFrame({'item_name':['macBook','Painting','Cup'], 'item_id':[0,1,2], 'cat_id':[0,5,2]})
b
```

```python
pd.merge(a,b, how='inner', on='cat_id')
```

### 2. outer join

this type opposite the inner

- the elemint, which not be same order -> will add -> NaN <-.



```python
pd.merge(a,b, how='outer', on='cat_id')
```

### 3. left join

will keep the elemint which in left side give it NaN
delet  elemint which in right.



```python
pd.merge(a,b, how='left', on='cat_id')
```

### 4. right
opposite of left

```python
pd.merge(a,b, how='right', on='cat_id')
```
