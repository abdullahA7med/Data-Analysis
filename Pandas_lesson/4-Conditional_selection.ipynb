**To solve it, need convert type of data in Stars To =>Float<=**

- **we use -> astype() function to convert to anouther dataType**

- **becouse some of Stars Values are string(here i'm not mean the type of theme, i mean there are string, have value => Unrated <=)**

```python
# 
# df['Stars'] > 2.5  | Error becouse type of data.
# df['Stars'].info() | info function, return some info about data, like type of dataType.

# df['Stars'].astype(float)  | there are string have values -> Unrated <-


# -----------------------------------------------------------------------
# ----> from here, i know the power of conditional selecting <-----------
# -----------------------------------------------------------------------

# 1. i will find the row which have "Unrated"
df['Stars'] == 'Unrated' # -> give me boolean value


# 2. (optional) if you need to see the rows which have unrated val
df.loc[df['Stars'] == 'Unrated']


# 3. now i need The data, that not equal 'Unrated'
df.loc[df['Stars'] != 'Unrated']


# 4. To save changes in to main dataFrame i do reassign. 
df = df.loc[df['Stars'] != 'Unrated']
df
```

**Now i fixed the second problem, removed rows that have 'Unrated' Value**

```python
# i did that changes just on -> Stars column <-

# so that i did -> reassign for Stars.

df['Stars'] = df['Stars'].astype(float)
```

```python

# To check if it Done.

df['Stars'].info()
```

```python
# 
# 
# Let's back to our question
#  get all rating that abover over average.


df.loc[df['Stars'] > 2.5]
```

6) **Get the not null values from the Top Ten column.**

**We have some fuction:**

- **isnull() => Return True if that value was null(NaN)**

..
    - **i can check all data OR specific column.**
..

- **notnull() => return True if that value was not null(NaN)**

```python

# now we see , we have many values have NaN.

df.head()
```

```python
df.isnull() # all Data
df['Top Ten'].isnull() # just check Top Ten column.
```

```python

# df['Top Ten'].isnull()     | -> returned boolean value.

# df[df['Top Ten'].isnull()] | -> return the data that have null in -> Top Ten <- column.


# we know this symbole -> ~ <- is not

df[~df['Top Ten'].isnull()] # | i need all values in > Top Ten < which not equal null (NaN).

```

```python

# The sec way.

# get all row from -> Top Ten <- column, which not equal null(NaN).

df[df['Top Ten'].notnull()]

# the same out put from first way.
```
