## 1. index based selection (implicit index)



................................................

- **Select the first row of the data**

```python
df.iloc[0,:] # return first row, with all column
```

..............
..............
............

- **Select the first column**

```python
df.iloc[:,0] # get all rows, with first column.
```

......................................
.................

- **Select the first 3 rows**


```python
df.iloc[:3, :]
```

........................................................................

- **Fancy indexing**

```python
# get row num: 1,4,10 | with all columns
df.iloc[[1,4,10],:]
```

..................................................................................

## Lable based selection(explicit index)

- **explicit index**

##### Select -> Brand , Style and Stars

```python
df.head()
```

```python
# select all rows to -> Brand, Style, Stars column

df.loc[:,['Brand','Style','Stars']]
```
