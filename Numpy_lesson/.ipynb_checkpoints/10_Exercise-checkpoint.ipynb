```python

# 8- what is maximum and minimum weight.

np.max(data[:,1]) , np.min(data[:,1])
```

```python

# mean = sumValues / numValues
# 9- find mean and standard deviation of height using the simple operation only

height = data[:,0]
# mean = np.sum(height) / 500
mean = np.sum(height) / height.size

print(mean)

diff = height - mean

squared = np.power(diff,2)
n = height.size
sm = np.sum(squared)

# this is standard devision.
np.sqrt(sm / (n - 1))
```
